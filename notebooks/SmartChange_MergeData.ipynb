```python
# Connection to the Vertica Database using ODBC.
import pyodbc
cur=pyodbc.connect("DSN=VerticaDSN").cursor()

# Import the Vertica ML Python library 
from vertica_ml_python import RVD
from vertica_ml_python import read_csv # This function will help us to load the csv file in the Database.
from vertica_ml_python import drop_table # This function will help us to drop the unnecessary tables.
```

```python
import pandas
import numpy as np
import sys
```

```python
# Load three data files
cm3rm1DF=pandas.read_csv('../datasets/cm3rm1.csv', encoding='windows-1252',dtype={'COLLISION_SEVERITY':np.str,'RISK_SEVERITY':np.str,'IMPACT_SEVERITY':np.str},converters = {'TIME_PERIOD_CONFLICT': lambda x: x if x else None})

cm4rm1DF=pandas.read_csv('../datasets/cm4rm1.csv')

cicm4rDF=pandas.read_csv('../datasets/cicm4r.csv')
```

```python
# merge 3 dataframes
ChangeDataDF=cm3rm1DF.merge(cm4rm1DF,on='NUMBER')

ChangeDataDF.merge(cicm4rDF,on='NUMBER')
```
